```python
import os
import shutil

# Directorio que contiene las imágenes originales
directorio_origen = "datathon/images"

# Directorio de destino donde se crearán las carpetas
directorio_destino_train = "datathon/color/train"

# Crea el directorio de destino si no existe
if not os.path.exists(directorio_destino_train):
    os.makedirs(directorio_destino_train)

# Directorio de destino donde se crearán las carpetas
directorio_destino_test = "datathon/color/test"

# Crea el directorio de destino si no existe
if not os.path.exists(directorio_destino_test):
    os.makedirs(directorio_destino_test)
    


# Itera a través de las imágenes en el directorio de origen
for file in range(0,7398):
    # Supongamos que quieres separar las imágenes en carpetas según su etiqueta
    filename = os.listdir(directorio_origen)[file]
    etiqueta = filename[-6:-4]  # Define una función para obtener la etiqueta

    # Crea la carpeta si no existe
    carpeta_etiqueta = os.path.join(directorio_destino_train, etiqueta)
    if not os.path.exists(carpeta_etiqueta):
        os.makedirs(carpeta_etiqueta)

    # Copia la imagen al directorio de destino
    ruta_origen = os.path.join(directorio_origen, filename)
    ruta_destino = os.path.join(carpeta_etiqueta, filename)
    shutil.copyfile(ruta_origen, ruta_destino)

    
# Itera a través de las imágenes en el directorio de origen
for file in range(7398,9248):
    # Supongamos que quieres separar las imágenes en carpetas según su etiqueta
    filename = os.listdir(directorio_origen)[file]
    etiqueta = filename[-6:-4]  # Define una función para obtener la etiqueta

    # Crea la carpeta si no existe
    carpeta_etiqueta = os.path.join(directorio_destino_test, etiqueta)
    if not os.path.exists(carpeta_etiqueta):
        os.makedirs(carpeta_etiqueta)

    # Copia la imagen al directorio de destino
    ruta_origen = os.path.join(directorio_origen, filename)
    ruta_destino = os.path.join(carpeta_etiqueta, filename)
    shutil.copyfile(ruta_origen, ruta_destino)
    
    
    
    
```

```python
# Initialising the CNN
color = Sequential()

# Step 1 - Convolution
color.add(Conv2D(32, (3, 3), input_shape = (334, 239, 3), activation = 'relu'))

# Step 2 - Pooling
color.add(MaxPooling2D(pool_size = (2, 2)))

# Adding a second convolutional layer
color.add(Conv2D(32, (3, 3), activation = 'relu'))
color.add(MaxPooling2D(pool_size = (2, 2)))

# Step 3 - Flattening
color.add(Flatten())

# Step 4 - Full connection
color.add(Dense(units = 254, activation = 'relu'))
color.add(Dense(units = 110, activation = 'softmax'))

# Compiling the CNN
color.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])




# Part 2 - Fitting the CNN to the images

from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

training_set = train_datagen.flow_from_directory('datathon/color/train',
                                                 target_size = (334, 239),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')

test_set = test_datagen.flow_from_directory('datathon/color/test',
                                            target_size = (334, 239),
                                            batch_size = 32,
                                            class_mode = 'categorical')

color.fit_generator(training_set,
                         steps_per_epoch = 128,
                         epochs = 100,validation_freq=0.2)
```

```python
color.save('./color_detection.h5')
```
